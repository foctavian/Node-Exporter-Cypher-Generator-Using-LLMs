```python
from pydantic import BaseModel, confloat, StringConstraints, Field, model_validator, field_validator
from typing import List, Optional, Dict, Union
import re

class Node(BaseModel):
  label: str
  properties: Dict[str, Union[str, int, float]]

  @field_validator('properties', mode='after')
  @classmethod
  def clean_invalid_backslashes(cls, value):
    cleaned_properties={}
    for k,v in value.items():
      if isinstance(v, str):
        cleaned_properties[k] = re.sub(r'\\', '', v)
      else:
        cleaned_properties[k] = v
    return cleaned_properties

  @staticmethod
  def is_balanced(text:str) -> bool:
    stack = []
    matching = {')': '(', '}': '{', ']': '['}
    for char in text:
      if char in "([{":
        stack.append(char)
      elif char in ')}]':
        if not stack or stack.pop != matching[char]:
          return False
    return not stack


  @model_validator(mode="before")
  @classmethod
  def validate_balanced_delimiters(cls, values: dict):
    label = values.get('label')
    if not cls.is_balanced(label):
      raise ValueError("Label is not balanced")

    properties = values.get('properties')
    for key, val in properties.items():
      if not cls.is_balanced(val):
        raise ValueError(f"Value for key '{key}' is not balanced")
      if isinstance(val, str) and not cls.is_balanced(val):
        raise ValueError(f"Unbalanced delimiters in property value for key `{key}`: {val}")

    return values



class NodeExporterMetrics(BaseModel):
  nodes: List[Node]
```

```python
cypher_keywords = [
    "CREATE", "MERGE", "MATCH", "OPTIONAL MATCH", "WHERE", "RETURN",
    "SET", "REMOVE", "DELETE", "DETACH DELETE", "FOREACH", "UNWIND",
    "WITH", "ORDER BY", "LIMIT", "SKIP"
]

def is_cypher_query(line: str):
  return any(keyword in line for keyword in cypher_keywords)
```

```python
from typing import List
import re

def clear_cypher_output(message: str) -> str:
    cleared_lines = []
    lines = message.split('\n')  # Split the content into individual lines.

    for i, line in enumerate(lines):
        stripped_line = line.strip()

        # # Skip comments (lines starting with `#` or containing `//`)
        # if stripped_line.startswith("#") or "//" in stripped_line:
        #     continue
        if not is_cypher_query(stripped_line):
          continue

        if stripped_line.startswith('`'):
          pattern=r'cypher'
          backticks_pattern = r"^`+"
          stripped_line = re.sub(backticks_pattern, "", stripped_line).strip()
          if re.search(pattern, stripped_line, re.IGNORECASE):
            continue
          else:
            stripped_line = re.sub(pattern, "", stripped_line).strip()

        # Skip lines with unbalanced parentheses or braces
        if stripped_line.count("(") != stripped_line.count(")") or stripped_line.count("{") != stripped_line.count("}"):
            continue

        # Handle cases where ON CREATE or ON MATCH are on a new line
        if stripped_line.startswith(("ON CREATE", "ON MATCH", "RETURN")):
            # Append to the previous line
            if cleared_lines:
                cleared_lines[-1] += f" {stripped_line}"
            continue

        # Add valid lines to the cleared list
        cleared_lines.append(stripped_line)

    # Join the cleared lines into a single string with line breaks
    return '\n'.join(cleared_lines)
```

```python
from json_repair import repair_json

def clear_json_llm_output(text: str) -> str:
  cleared_lines=[]
  lines = text.split('\n')

  for i, line in enumerate(lines):
    stripped_line = line.strip()

    if stripped_line.startswith('`'):
      continue
    cleared_lines.append(stripped_line)

  json = '\n'.join(cleared_lines)

  return repair_json(json)


```

```python
def split_lines_into_chunks(lines, max_tokens=8192):
    chunks = []
    current_chunk = []
    current_token_count = 0

    for line in lines:
        # Estimate token count per line (tokenization is needed for more accuracy)
        tokens_in_line = len(line.split())  # This is a simple approximation
        if current_token_count + tokens_in_line <= max_tokens:
            current_chunk.append(line)
            current_token_count += tokens_in_line
        else:
            # If the chunk exceeds max tokens, save the current chunk and start a new one
            chunks.append("\n".join(current_chunk))
            current_chunk = [line]
            current_token_count = tokens_in_line

    # Add the last chunk if there is any leftover data
    if current_chunk:
        chunks.append("\n".join(current_chunk))

    return chunks
```
